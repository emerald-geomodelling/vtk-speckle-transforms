A short example notebook demonstrating how to open up and access the files in `data/example_project_01` pythonically using pyvista. 

# Import statements

```python
import pyvista as pv
import numpy as np
```

# User parameters

```python
files_in = {
    'aem':'../data/example_project_01/AEM-resistivity-model-compressed.vtu',
    'bedrock':'../data/example_project_01/bedrock-model-ANN.vtu',
    'quickclay':'../data/example_project_01/quick-clay-probability-compressed.vtu',
}
```

# load data

```python
ugrids = {}
for name, path in files_in.items():
    print(f'reading {name} at location {path}')
    ugrids[name]=pv.read(path)
```

# preview each file's contents
print both a summary of the data arrays, and show at least one interesting array

```python
ugrids['aem'].plot(scalars='resist',log_scale=True, clim=[1,2000], cmap='magma')
```
